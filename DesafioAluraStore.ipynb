```python
import pandas as pd

# Carregando os arquivos CSV
loja_1 = pd.read_csv("loja_1.csv")
loja_2 = pd.read_csv("loja_2.csv")
loja_3 = pd.read_csv("loja_3.csv")
loja_4 = pd.read_csv("loja_4.csv")

# Visualizando as primeiras linhas de cada loja
print("Loja 1")
print(loja_1.head(), "\n")

print("Loja 2")
print(loja_2.head(), "\n")

print("Loja 3")
print(loja_3.head(), "\n")

print("Loja 4")
print(loja_4.head())
```

```python
# Calculando o faturamento total de cada loja (preço + frete)
faturamento_lojas = {
    "Loja 1": (loja_1["Preço"] + loja_1["Frete"]).sum(),
    "Loja 2": (loja_2["Preço"] + loja_2["Frete"]).sum(),
    "Loja 3": (loja_3["Preço"] + loja_3["Frete"]).sum(),
    "Loja 4": (loja_4["Preço"] + loja_4["Frete"]).sum(),
}

# Exibindo os resultados
for loja, faturamento in faturamento_lojas.items():
    print(f"{loja}: R$ {faturamento:,.2f}")
```

```python
# Função para contar as categorias mais populares
def categorias_populares(df, nome_loja):
    print(f"\n{nome_loja} - Categorias mais vendidas:")
    print(df["Categoria do Produto"].value_counts().head())

# Análise por loja
categorias_populares(loja_1, "Loja 1")
categorias_populares(loja_2, "Loja 2")
categorias_populares(loja_3, "Loja 3")
categorias_populares(loja_4, "Loja 4")

```

```python
# Cálculo da média de avaliação por loja
avaliacoes_medias = {
    "Loja 1": loja_1["Avaliação da compra"].mean(),
    "Loja 2": loja_2["Avaliação da compra"].mean(),
    "Loja 3": loja_3["Avaliação da compra"].mean(),
    "Loja 4": loja_4["Avaliação da compra"].mean(),
}

# Exibindo os resultados
for loja, media in avaliacoes_medias.items():
    print(f"{loja}: média de avaliação = {media:.2f}")

```

```python
def produtos_extremos(df, nome_loja):
    print(f"\n{nome_loja}")
    print("Mais vendidos:")
    print(df["Produto"].value_counts().head(3))
    print("Menos vendidos:")
    print(df["Produto"].value_counts().tail(3))

produtos_extremos(loja_1, "Loja 1")
produtos_extremos(loja_2, "Loja 2")
produtos_extremos(loja_3, "Loja 3")
produtos_extremos(loja_4, "Loja 4")
```

```python
# Cálculo do frete médio
frete_medio = {
    "Loja 1": loja_1["Frete"].mean(),
    "Loja 2": loja_2["Frete"].mean(),
    "Loja 3": loja_3["Frete"].mean(),
    "Loja 4": loja_4["Frete"].mean(),
}

# Exibir resultado
for loja, frete in frete_medio.items():
    print(f"{loja}: frete médio = R$ {frete:.2f}")
```

```python
import matplotlib.pyplot as plt

# Dados
lojas = list(faturamento_lojas.keys())
valores = list(faturamento_lojas.values())

# Gráfico
plt.figure(figsize=(8,5))
plt.bar(lojas, valores, color='skyblue')
plt.title("Faturamento Total por Loja")
plt.xlabel("Lojas")
plt.ylabel("Faturamento (R$)")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# Dados
avaliacoes = list(avaliacoes_medias.values())

# Gráfico
plt.figure(figsize=(8,5))
plt.plot(lojas, avaliacoes, marker='o', color='orange')
plt.title("Média de Avaliação por Loja")
plt.xlabel("Lojas")
plt.ylabel("Avaliação Média")
plt.ylim(3.8, 4.2)
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
# Dados
fretes = list(frete_medio.values())

# Gráfico
plt.figure(figsize=(7,7))
plt.pie(fretes, labels=lojas, autopct="%1.1f%%", colors=['#FF9999','#66B3FF','#99FF99','#FFD700'])
plt.title("Distribuição do Frete Médio por Loja")
plt.tight_layout()
plt.show()
```

## Métricas Analisadas

### 1. Faturamento Total
A Loja 4 apresentou o menor faturamento entre todas as unidades, totalizando **1.458.253,46 Reais.** Já a Loja 1 obteve o maior faturamento, com **1.616.347,09 Reais.**

### 2. Categorias Mais Vendidas
As categorias com maior volume de vendas foram similares entre as lojas, com destaque para **móveis** e **eletrônicos**. A Loja 2 apresentou o menor volume nas 5 categorias mais vendidas.

A Loja 4 possui o menor frete médio (**R$ 31,28**), o que é um ponto positivo. Porém, esse fator não compensa seu baixo faturamento.

Com base nas métricas analisadas, a **Loja 4 apresenta o pior desempenho geral**, especialmente em **faturamento total**, o que é um indicativo forte de baixo retorno financeiro. Apesar de seu frete ser o mais barato, isso não se traduz em maior volume de vendas nem em melhores avaliações.
